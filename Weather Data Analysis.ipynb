# 📊  Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Clean column names for easier access
data.columns = data.columns.str.strip().str.replace(' ', '_').str.replace('%', 'pct')
```

```python
# Data overview
data.info()
```

```python
# Check for null values
data.isnull().sum()
```

```python
# Convert Date/Time column to datetime format
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

# Temperature Over Time

```python
plt.figure(figsize=(15,5))
plt.plot(data['Date/Time'], data['Temp_C'], color='orange')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

# Humidity vs Temperature

```python
sns.scatterplot(x='Temp_C', y='Rel_Hum_pct', data=data, alpha=0.5)
plt.title('Humidity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.show()
```

```python
# Weather Condition Frequency
print(data.columns)
```

```python
data['Weather_Condition'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Weather Conditions')
plt.xlabel('Frequency')
plt.gca().invert_yaxis()
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Weather Condition Count

```python
data['Weather_Condition'].value_counts().head(10).plot(kind='barh')
```

# 📈 1. Average Temperature by Month

```python
data['Month'] = data['Date/Time'].dt.month
monthly_avg_temp = data.groupby('Month')['Temp_C'].mean()
monthly_avg_temp.plot(kind='bar', color='salmon')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Avg Temperature (°C)')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 🌧 2. Days with Fog, Rain, or Snow

```python
fog_days = data[data['Weather_Condition'].str.contains('Fog', case=False)].shape[0]
rain_days = data[data['Weather_Condition'].str.contains('Rain|Showers|Drizzle', case=False)].shape[0]
snow_days = data[data['Weather_Condition'].str.contains('Snow', case=False)].shape[0]

conditions = ['Fog', 'Rain', 'Snow']
counts = [fog_days, rain_days, snow_days]

plt.bar(conditions, counts, color=['gray', 'blue', 'lightblue'])
plt.title('Days with Fog, Rain, or Snow')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```
